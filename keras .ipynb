```python
import tensorflow as tf
from tensorflow import keras
import numpy as np

  
    
```

```python
a = np.array([5,10,15,12])
b= np.array([15,25,35,29])
model = keras.Sequential()
model.add(keras.layers.Dense(1,input_shape=[1]))
model.compile(optimizer = tf.keras.optimizers.Adam(.8) , loss = 'mean_squared_error',metrics = ['accuracy'])
model.fit(a,b,epochs = 100)
pred = model.predict([100,1000,100000000])
```
