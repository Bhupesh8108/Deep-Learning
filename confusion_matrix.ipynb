```python
import tensorflow as tf 
import numpy as np 
import seaborn as se 
import matplotlib.pyplot as plt
from tensorflow import keras
```

```python
(x,y),(xt,yt) = keras.datasets.mnist.load_data()
x = x.reshape(len(x),28*28)
model = keras.Sequential([
    keras.layers.Dense(100,input_shape = (28**2,) ,activation = 'relu'),
                         keras.layers.Dense(10,activation = 'sigmoid')])
model.compile(optimizer = 'adam' , loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
model.fit(x,y,epochs=20)
xt = xt.reshape(len(xt),28**2)
pred = model.predict(xt)
pred = [np.argmax(i) for i in pred]
```

```python
cm = tf.math.confusion_matrix(labels = yt,predictions = pred)
se.heatmap(cm,annot = True ,fmt='d')
```
