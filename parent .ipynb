```python
import pandas as pd
data = pd.read_csv('/home/kali/Downloads/lego-analysis-master/datasets/parent_themes.csv')
for i in range (len(data)):
    old = data['name'].unique()[i]
    data = data.replace({old:i})
data.head(50)
```

```python
import tensorflow as ten
from tensorflow import keras
import numpy as np


model = keras.Sequential()
model.add(keras.layers.Flatten(input_shape = [1]))
model.add(keras.layers.Dense(10,activation = 'relu'))
model.add(keras.layers.Dense(2,activation = 'sigmoid'))
model.compile(optimizer = 'adam' , metrics = 'accuracy',loss = 'sparse_categorical_crossentropy')
model.fit(data['name'] , data['is_licensed'],epochs = 10)
pred = model.predict(data['name'])
```

```python
pred = [np.argmax(i) for i in pred]
from collections import Counter
cout = Counter()
count =cout.update(Counter(pred))
count
```
