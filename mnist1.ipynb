```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt

```

```python
(x,y),(xt,yt) = keras.datasets.mnist.load_data()
# plt.matshow(x[0])
plt.imshow(x[0])
xt.shape
```

```python
# x = x/255
# xt = xt/255
x = x.reshape(len(x),28*28)
xt = xt.reshape(len(xt),28*28)
model = keras.Sequential([
    keras.layers.Flatten(input_shape = (28**2,)),
    keras.layers.Dense(100,activation = 'relu'),
keras.layers.Dense(10,activation = 'sigmoid')])
model.compile(optimizer = 'adam' ,loss = 'sparse_categorical_crossentropy',metrics = ['accuracy'])
model.fit(x,y,epochs =5)
# model.evaluate(x,y)
pred = model.predict(xt)
# keras.math.cofusion_matrix
pred = [np.argmax(i) for i in pred]
cm = tf.math.confusion_matrix(labels=yt,predictions = pred)

```

```python
import seaborn as se
se.heatmap(cm,annot = True,fmt = 'd')
model.evaluate(x,y)
```
